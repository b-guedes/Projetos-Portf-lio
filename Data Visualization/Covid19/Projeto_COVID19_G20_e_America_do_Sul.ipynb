# **1. INTRODUÇÃO**

Este projeto têm por intuito analisar dados da pandemia de COVID-19 com o auxílio de ferramentas de visualização de dados - [Power BI](https://powerbi.microsoft.com/) e [Looker Studio](https://lookerstudio.google.com/).

É concesso hoje que o trabalho de monitoramento estatístico feito durante a pandemia foi uma importante ferramenta para atuação dos governos e da sociedade no enfrentamento a pandemia da COVID-19[¹](https://www2.ufjf.br/noticias/2020/04/22/pesquisador-explica-como-a-estatistica-ajuda-no-combate-a-covid-19/#:~:text=Em%20tempos%20de%20crise%2C%20como,por%20exemplo%2C%20o%20isolamento%20social%2C) [²](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC8577168/) [³](https://mitsloan.mit.edu/ideas-made-to-matter/how-covid-19-disrupting-data-analytics-strategies). Graças às ferramentas que trabalham com dados disponíveis hoje, a pandemia, que foi brutal e devastadora, pode ser controlada com certa rapidez e, nos locais onde as autoridades e população trabalharam em conjuntos para mitigar a contaminação, os estragos sociais foram menores do que se previa no início do ocorrido. É muito em consideração do trabalho que essas tecnologias proporcionaram que hoje, 4 anos após o início, o senso de normalidade tenha voltado a maioria da população global.

Com esse contexto, buscarei neste projeto trabalhar um pouco os dados coletados, trabalhados e melhorados de forma exaustiva por muitos pesquisadores e instituições. Dessa forma, espero demonstrar alguns parâmetros gerais de como foram esses anos de pandemia da COVID-19.

Neste bloco encontram-se países de: todos os continentes; 7 dos 10 mais populosos; 8 dos 10 de maior área e; as 10 maiores economias do mundo. Assim sendo, o peso do Grupo é imenso para medir qualquer parâmetro no mundo atual. Desta forma, espero poder demonstrar de forma global como foi o comportamento da COVID-19 até o momento.

## 2.1 Explorando o Dataset

Com os dados explicitados e explicados, vamos ao código e a montagem do conjunto de dados que irei utilizar.

```python
#importar as bibliotecas que serão utilizadas
import requests


import numpy as np
import pandas as pd


import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#abrindo o arquivo no Pandas
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'

df_casos = pd.read_csv(url, sep=',')
```

Vamos olhar os dados contidos no arquivo CSV que está disponível, assim será decidido quais dados precisam de alterações e como elas serão feitas.


Além de uma visão rápida na tabela, também será verificado alguns dados mais gerais dos dados como o tipo, tamanho das variáveis, etc.

Os dados sofreram algumas alterações ao longo do tempo, no entanto, há algum tempo foi encontrado um formato muito bem aceito para análises da COVID-19 como, por exemplo, acrescido de níveis de suavização dos dados diários para maior precisão.

Dessa forma, por se tratar de um material muito bem trabalhado pela **OWID**, as correções e alterações devem ser poucas, vamos focar em limpar o conjunto de dados para atender melhor as necessidades e, também, melhorar o desempenho. Assim, será dividido em dois conjuntos -dos países do G20 e da América do Sul, conforme já explicitado anteriormente.

```python
#retirando as colunas que não interessam ao projeto
df_limpando = df_casos.drop(['icu_patients',
                             'icu_patients_per_million',
                             'hosp_patients',
                             'hosp_patients_per_million',
                             'weekly_icu_admissions',
                             'weekly_icu_admissions_per_million',
                             'weekly_hosp_admissions',
                             'weekly_hosp_admissions_per_million',
                             'reproduction_rate',
                             'total_tests',
                             'new_tests',
                             'total_tests_per_thousand',
                             'new_tests_per_thousand',
                             'new_tests_smoothed',
                             'new_tests_smoothed_per_thousand',
                             'positive_rate',
                             'tests_per_case',
                             'tests_units',
                             'extreme_poverty',
                             'population_density',
                             'stringency_index',
                             'aged_65_older',
                             'aged_70_older',
                             'total_vaccinations_per_hundred',
                             'people_vaccinated_per_hundred',
                             'people_fully_vaccinated_per_hundred',
                             'total_boosters_per_hundred',
                             'new_people_vaccinated_smoothed_per_hundred',
                             'cardiovasc_death_rate',
                             'diabetes_prevalence',
                             'female_smokers',
                             'male_smokers',
                             'handwashing_facilities',
                             'iso_code'
                             ], axis=1)
```

Observando como ficou o Dataset após a eliminação dessas colunas:

```python
df_limpando.query('continent == "Asia" & excess_mortality_cumulative > 0').tail()
```

Irei trocar alguns tipos de objetos para garantir uma maior qualidade aos números no momento da análise.

Primeiro é necessário garantir que todos os valores NaN (nulos) estejam preenchidos. A minha escolha foi por substituir todos por 0, pois assim não será gerada distorções nos dados, como alguns dos outros métodos poderiam gerar.

```python
#preenchendo os dados vazios
df_limpando = df_limpando.fillna(0)
```

```python
#garantindo os tipos de cada coluna - int e datetime
df_limpando['total_cases'] = df_limpando['total_cases'].astype('int64')
df_limpando['new_cases'] = df_limpando['new_cases'].astype('int64')
df_limpando['total_deaths'] = df_limpando['total_deaths'].astype('int64')
df_limpando['new_deaths'] = df_limpando['new_deaths'].astype('int64')
df_limpando['total_vaccinations'] = df_limpando['total_vaccinations'].astype('int64')
df_limpando['people_vaccinated'] = df_limpando['people_vaccinated'].astype('int64')
df_limpando['people_fully_vaccinated'] = df_limpando['people_fully_vaccinated'].astype('int64')
df_limpando['total_boosters'] = df_limpando['total_boosters'].astype('int64')
df_limpando['new_vaccinations'] = df_limpando['new_vaccinations'].astype('int64')
df_limpando['date'] = df_limpando['date'].astype('datetime64[ns]')
df_limpando['population'] = df_limpando['population'].astype('int64')
```

```python
#garantindo os tipos de cada coluna - float - e o arredondamento correto
df_limpando['new_cases_smoothed'] = df_limpando['new_cases_smoothed'].astype('float64').round(3)
df_limpando['new_deaths_smoothed'] = df_limpando['new_deaths_smoothed'].astype('float64').round(3)
df_limpando['total_cases_per_million'] = df_limpando['total_cases_per_million'].astype('float64').round(3)
df_limpando['new_cases_per_million'] = df_limpando['new_cases_per_million'].astype('float64').round(3)
df_limpando['new_cases_smoothed_per_million'] = df_limpando['new_cases_smoothed_per_million'].astype('float64').round(3)
df_limpando['total_deaths_per_million'] = df_limpando['total_deaths_per_million'].astype('float64').round(3)
df_limpando['new_deaths_per_million'] = df_limpando['new_deaths_per_million'].astype('float64').round(3)
df_limpando['new_deaths_smoothed_per_million'] = df_limpando['new_deaths_smoothed_per_million'].astype('float64').round(3)
df_limpando['new_vaccinations_smoothed'] = df_limpando['new_vaccinations_smoothed'].astype('float64').round(3)
df_limpando['new_vaccinations_smoothed_per_million'] = df_limpando['new_vaccinations_smoothed_per_million'].astype('float64').round(3)
df_limpando['new_people_vaccinated_smoothed'] = df_limpando['new_people_vaccinated_smoothed'].astype('float64').round(3)
df_limpando['gdp_per_capita'] = df_limpando['gdp_per_capita'].astype('float64').round(3)
df_limpando['hospital_beds_per_thousand'] = df_limpando['hospital_beds_per_thousand'].astype('float64').round(3)
df_limpando['life_expectancy'] = df_limpando['life_expectancy'].astype('float64').round(2)
df_limpando['human_development_index'] = df_limpando['human_development_index'].astype('float64').round(3)
df_limpando['excess_mortality_cumulative_absolute'] = df_limpando['excess_mortality_cumulative_absolute'].astype('float64').round(3)
df_limpando['excess_mortality_cumulative'] = df_limpando['excess_mortality_cumulative'].astype('float64').round(3)
df_limpando['excess_mortality'] = df_limpando['excess_mortality'].astype('float64').round(3)
df_limpando['excess_mortality_cumulative_per_million'] = df_limpando['excess_mortality_cumulative_per_million'].astype('float64').round(4)
```

Devemos levar em consideração que se trata de uma pandemia, então é de se esperar alguns casos de exceção com números fora do padrão, também denominados *outliers*.

Abaixo, utilizarei o pacote [Seaborn](https://seaborn.pydata.org/) para extrair o *Boxplot* (Diagrama de caixa), uma ferramenta útil para extrair os dados de forma descritiva a fim de, através de quartis, entender uma variável numérica. Esta ferramenta estatística deixa muito evidente se existe "pontos fora da curva" - denominados de *outliers*.

Vamos poder analisar um exemplo. Utilizando a categoria "*new_deaths_smoothed*" podemos ver como o número de mortos possui diversos outliers, o que poderia indicar alguma inconsistência nos dados.

No entanto, é fácil lembrar que esses números altos de fato existiram e, então, os dados estão provavelmente corretos - por exemplo, no Brasil a média móvel com mais de 3 mil casos é relatada nestas reportagens [1](https://g1.globo.com/bemestar/coronavirus/noticia/2021/04/12/brasil-volta-a-bater-pior-media-movel-de-mortes-por-covid-com-3125-obitos-por-dia-na-ultima-semana.ghtml),[2](https://www.cnnbrasil.com.br/saude/pelo-terceiro-dia-consecutivo-o-brasil-registra-mais-de-3700-mortes/) e no Peru também são relatadas por essas reportagens [3](https://veja.abril.com.br/mundo/covid-19-como-peru-se-tornou-pais-com-maior-taxa-de-mortalidade-do-mundo/), [4](https://www.npr.org/sections/goatsandsoda/2021/11/27/1057387896/peru-has-the-worlds-highest-covid-death-rate-heres-why).
