# 1. **INTRODUÇÃO**

Este projeto procura analisar o Dataset da empresa **Delivery Center** realizando a **Análise Exploratória dos Dados (EDA)**. 

A **Delivery Center** foi uma **plataforma de logística e entregas** que tinha como diferencial o atendimento a clientes instalados em shopping centers de grandes capitais do Brasil. Os **clientes** eram tanto lojas do **ramo alimentício** (denominados como *food* no dataset) quanto dos segmentos de **bens de consumo** (aqui denominados como *good*) 

Dessa forma, suas centrais de distribuição (***hubs***) ficavam localizadas em shoppings do Brasil e, neste conjunto de dados, são fornecidos dados de **Curitiba**, **Porto Alegre**, **Rio de Janeiro** e **São Paulo**. A estrutura dos dados fornecidas são semelhantes aos dados reais que eram utilizados pela empresa. 

A empresa obteve relativo sucesso durante sua trajetória - *fundada em 2016* ela teve suas operações *encerradas em 2021* - conseguindo aportes financeiros importantes de grandes grupos do setor de centros comerciais no Brasil como a **BR Mall** e a **Multiplan** que, eventualmente, se tornaram, também, acionistas da empresa. 

Ao longo dessa trajetória chamou a atenção desse segmento de mercado uma nova forma de conectar lojistas de centros urbanos ao e-commerce, trazendo a possibilidade dessas empresas de aproveitar os estoques desses locais para aumentar as vendas e diminuir os custos de distribuição e logística. 

Durante os anos do auge da pandemia da *COVID-19* (entre os anos 2019-2020) a empresa obteve crescimento significativo, o que fez dela cobiçada e alvo de grande interesse por parte do mercado. Em [determinado](https://www.mobiletime.com.br/noticias/28/01/2019/delivery-center-instala-hubs-para-entregas-rapidas-em-shopping-centers/) momento chegou a agregar a logística, concomitante, de aplicativos com foco em produtos de alimentação como iFood, Uber Eats, Rappi - concorrentes entre si, e tinha como objetivo ganhar mais terreno em operações no segmento de bens de consumo onde empresas como Amazon, Magazine Luiza e Mercado Livre são consolidadas. 

A [grande oportunidade](https://oglobo.globo.com/economia/emprego/a-forca-do-delivery-shoppings-centers-fazem-parcerias-para-se-tornarem-centros-de-distribuicao-24061415) foi vista em se tornar uma referência no mercado denominado como ["last mile"](https://www.totvs.com/blog/gestao-logistica/last-mile/) - a última etapa do processo de logística onde o produto chega ao cliente final. No fim, a empresa enfrentou um mercado que cresceu rapidamente e teve uma grande concorrência, seguido de [conflitos](https://valor.globo.com/empresas/noticia/2021/11/23/apos-investirem-r-180-milhoes-socios-fecham-delivery-center.ghtml) entre os acionistas e as necessidades da empresa, uma queda de faturamento e falta de perspectiva futuras a curto prazo, culminaram com o fechamento da empresa no final de 2021.

### Carregar os pacotes

```python
#carrega os pacotes utilizados

from termcolor import colored
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def missing_data(data):
    """
    Calculates information about missing data in a DataFrame.

    Args:
        data (pd.DataFrame): The input DataFrame.

    Returns:
        pd.DataFrame: A summary of missing data including total count, unique values,
                      null count, and percentage of missing values.
    """
    total = data.count()
    unique = data.nunique()
    null = data.isnull().sum()
    percent = round((data.isnull().sum() / data.isnull().count() * 100), 2)
    tt = pd.concat([total, unique, null, percent], axis=1, keys=['Total', 'Unique', 'Null', 'Percent'])
    types = []
    for col in data.columns:
        dtype = str(data[col].dtype)
        types.append(dtype)
    tt['Types'] = types
    return tt.transpose()


def count_by_column(df, count_column):
    """
    Counts occurrences of values and calculates the percentage in the specified count_column.

    Args:
        df (pandas.DataFrame): The input DataFrame.
        count_column (str): Column name for counting occurrences.

    Returns:
        pandas.DataFrame: A DataFrame with count of occurrences and percentage.
    """
    count = df[count_column].value_counts().reset_index()
    count.columns = [count_column, 'occurrences']  # Rename the columns for clarity
    count['percentage'] = round(100 * (count['occurrences'] / count['occurrences'].sum()), 2)
    return count



def freq_column(df, column):
    """
    Calculate the frequency distribution of a column in a DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column (str): The name of the column to analyze.

    Returns:
        pd.Series: A Series containing the frequency distribution.
    """
    n = len(df) #df size
    k = np.ceil(1 + 3.322 * np.log10(n)) #number of classes
    sorted_data = np.sort(df[column]) #sort value ascending
    data_range = np.max(sorted_data) - np.min(sorted_data) #range of df
    bin_width = np.ceil(data_range / k) #range of classes
    bins = [np.min(sorted_data) + i * bin_width for i in range(int(k) + 1)] #limits of classes
    freq = pd.cut(sorted_data, bins=bins, right=False).value_counts() #frequency's table
    freq_df = pd.DataFrame(freq)
    freq_df["Percentage"] = round((freq_df["count"] / n) * 100, 2)
    return freq_df
```

## 3.2.1 Outliers

Um modo bastante comum de visualizar outliers é através do gráfico da caixa ou *boxplot*. Nele fica muito claro visualmente onde estão os outliers e o quão distante eles se encontram em relação ao conjunto geral. 

Abaixo colocarei as 4 principais categorias com *outliers* encontradas anteriormente neste tipo de gráfico, elas são: **order_amount, delivery_distance_meters, order_delivery_fee e order_delivery_cost**. Ficarão claro como os outliers presentes estão muito fora das caixas, quase impossibilitando de ter uma visualização gráfica dela.

É importante destacar que o tratamento desses dados não significa que eles estejam incorretos, para ter certeza disto teria de ser feita uma análise junto à empresa, no entanto, esses dados representam situações extremas que atrapalham a visualização das medidas resumo e, também, a possível utilização em modelos de treinamento de máquina. Por essas razões é que optei por fazer esse tratamento desses dados como seguira.

```python
#cria plot para mais de uma figura
fig, axs = plt.subplots(2, 2,figsize=(6, 6),layout='constrained')

#cria os gráficos de boxplot
axs[0, 0].boxplot(df_todos["order_amount"])
axs[0, 0].set_title("Order Amount")

axs[0, 1].boxplot(df_todos["delivery_distance_meters"])
axs[0, 1].set_title("Delivery Distance (m)")

axs[1, 0].boxplot(df_todos["order_delivery_fee"])
axs[1, 0].set_title("Delivery Fee")

axs[1, 1] = sns.boxplot(df_todos["order_delivery_cost"])
axs[1, 1].set_title("Delivery Cost")

plt.show()
```

Para fazer o tratamento dos *outliers*, optei por fazer o cálculo da Amplitude Interquartil (**IQR**) de cada variável de interesse (ou seja, o tamanho da *caixa* no boxplot). Assim, é possível determinar quais valores são outliers e, por tanto, devem ser excluídos do conjunto de dados.

```python
Q1 = df_todos.select_dtypes(include=['float64', 'int64']).quantile(0.25)
Q3 = df_todos.select_dtypes(include=['float64', 'int64']).quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

Sabendo a IQR de cada variável é possível excluir os valores que estejam fora dela dentro dos datasets. A seguir coloco, como forma de comparar, o tamanho dos 3 datasets calculados: **a)** com outliers, **b)** sem outliers extremos e **c)** sem outliers extremos e moderados.

```python
df = df_todos.select_dtypes(include=['float64', 'int64'])
df1 = df[~((df < (Q1 - 3 * IQR)) |(df > (Q3 + 3 * IQR))).any(axis=1)] #extremo
df2 = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)] # moderado e extremo 
print(f'Sem outliers moderados e extremos: {df2.shape}')
print(f'Sem outliers extremos: {df1.shape}')
print(f'Com outliers: {df_todos.shape}')
```

Para deixar ainda mais explícito o que foi feito colocarei a seguir os dados trabalhados em representações gráficas de histograma de cada variável. 

A primeira representação, em **vermelho**, mostra um conjunto muito desigual, isso é resultado das quantias extremas que acaba por gerar uma análise imprecisa e/ou distorcida dos dados. 

As representações em *amarelo* e *verde* parecem ser outra realidade, tamanha a diferença! Em **amarelo** estão os dados sem outliers extremos, a mudança já é significativa em relação ao conjunto original, podemos notar como existem padrões visíveis de comportamento nesses dados. Mas eles ainda podem ser melhores, e é isso que está demonstrado nas representações em **verde**, este representa o conjunto sem os outliers extremos e moderados, dessa maneira a confiabilidade do conjunto fica ainda mais alta.

Para ter uma noção mais clara da modificação, calculo as medidas de assimetria (skewness) e curtose (kurtosis) das curvas dos dados em cada uma das situações tomando como exemplo a variável *order_amount*.

```python
# Define a function to plot histograms
def plot_histograms(df, title, color):
    df.select_dtypes(include=['float64']).hist(edgecolor='black', color=color, bins=15, figsize=(10, 5), layout=(2, 3))
    plt.suptitle(title)

# Plot histograms for the original dataset (with no alterations)
plot_histograms(df_todos, 'Histograms with Outliers', 'red')

# Plot histograms for the dataset without extreme outliers
plot_histograms(df1, 'Histograms without Extreme Outliers', 'yellow')

# Plot histograms for the dataset without moderate and extreme outliers
plot_histograms(df2, 'Histograms without Extreme and Moderate Outliers', 'green')

plt.show()
```

```python
print("Skewness com outliers: {:0.3f}".format(df_todos["order_amount"].skew()))
print("Skewness sem outliers extremos: {:0.3f}".format(df1["order_amount"].skew()))
print("Skewness sem outliers moderados e extremos: {:0.3f}".format(df2["order_amount"].skew()))
print("Kurtosis com outliers: {:0.3f}".format(df_todos["order_amount"].kurt()))
print("Kurtosis sem outliers extremos: {:0.3f}".format(df1["order_amount"].kurt()))
print("Kurtosis sem outliers moderados e extremos: {:0.3f}".format(df2["order_amount"].kurt()))
```
